```python
import numpy as np
import cv2
import sys
import os
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
current_dir = Path(os.getcwd())
image_path = current_dir.parent / "images" / "kbskl.jpeg"
```

```python
image = cv2.imread(str(image_path), cv2.IMREAD_COLOR)
```

```python
if image is None:
    sys.exit("Could not read the image.")
```

```python
res = cv2.resize(image, None, fx=0.3, fy=0.3)
res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
```

```python
plt.title("Resize + Convert")
plt.imshow(res)
plt.axis("off")  
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(res, (5,5), 0)
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.subplot (1,2,1)
plt.imshow(blurred_image)
plt.title("Gaussian")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(res)
plt.title("Original")
plt.axis("off")
```
