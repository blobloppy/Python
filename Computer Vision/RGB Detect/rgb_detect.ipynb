```python
import cv2
import numpy as np

vid = cv2.VideoCapture(0)  # Change 0 to the video file path if using a file

while True:
    ret, frame = vid.read()  # 'ret' indicates if frame was successfully captured

    if not ret or frame is None:
        print("Error: Could not read frame. Exiting...")
        break  # Exit the loop if no frame is captured

    cv2.imshow("frame", frame)

    # Extract color channels safely
    red = frame[:, :, 2]  # OpenCV uses BGR format
    green = frame[:, :, 1]
    blue = frame[:, :, 0]

    # Compute mean values
    red_mean = np.mean(red)
    green_mean = np.mean(green)
    blue_mean = np.mean(blue)

    # Determine the dominant color
    if blue_mean > green_mean and blue_mean > red_mean:
        print("Blue")
    elif green_mean > blue_mean and green_mean > red_mean:
        print("Green")
    else:
        print("Red")

    # Break loop on 'q' press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
vid.release()
cv2.destroyAllWindows()
```
