```python
import numpy as np
import scipy.ndimage
import matplotlib.pyplot as plt

# Define the 10x10 pixel table (grayscale intensity values)
pixel_table = np.array([
    [25, 25, 25, 25, 25, 25, 25, 25, 25],
    [25, 25, 25, 25, 25, 25, 25, 25, 25],
    [25, 25, 230, 230, 25, 230, 230, 25, 25],
    [25, 25, 230, 230, 25, 230, 230, 25, 25],
    [25, 25, 230, 230, 230, 230, 230, 25, 25],
    [25, 25, 230, 230, 25, 230, 230, 25, 25],
    [25, 25, 230, 230, 25, 230, 230, 25, 25],
    [25, 25, 230, 230, 25, 230, 230, 25, 25],
    [25, 25, 25, 25, 25, 25, 25, 25, 25],
    [25, 25, 25, 25, 25, 25, 25, 25, 25]
], dtype=np.float32)

# Define a 3x3 Gaussian kernel (approximation)
gaussian_kernel = np.array([
    [0.09, 0.11, 0.09],
    [0.11, 0.13, 0.11],
    [0.09, 0.11, 0.09]
], dtype=np.float32)

# Normalize kernel (ensuring sum = 1)
gaussian_kernel /= gaussian_kernel.sum()

# Apply Gaussian filter using convolution
blurred_pixels = scipy.ndimage.convolve(pixel_table, gaussian_kernel, mode='constant')

# Display the original and blurred pixel tables
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(pixel_table, cmap='gray', vmin=0, vmax=255)
axes[0].set_title("Original Pixel Table")
axes[0].axis("off")

axes[1].imshow(blurred_pixels, cmap='gray', vmin=0, vmax=255)
axes[1].set_title("Blurred Pixel Table")
axes[1].axis("off")

plt.show()

# Print the result as a matrix
print("Blurred Pixel Table:")
print(np.round(blurred_pixels, 2))  # Rounded for better readability
```
