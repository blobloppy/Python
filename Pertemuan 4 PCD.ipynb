```python
from PIL import Image, ImageFilter
```

```python
im = Image.open('images\\kucing.jpg')
```

```python
im
```

```python
im_pixel = im.load()
```

```python
lebar = im.size[0]
tinggi = im.size[1]
grayscale = Image.new(mode="RGB", size=(lebar, tinggi))
gray_pixel = grayscale.load()
```

```python
for x in range(0, lebar):
    for y in range(0, tinggi):
        red = im_pixel[x,y][0]
        green = im_pixel[x,y][1]
        blue = im_pixel[x,y][2]
        gray = int((red+green+blue)/3)
        gray_pixel[x,y] = (gray, gray, gray)
```

```python
grayscale
```

```python
inverse = Image.new(mode="RGB", size=(lebar, tinggi))
inverse_pixel = inverse.load()
for x in range(0, lebar):
    for y in range(0, tinggi):
        red = 255 - im_pixel[x,y][0]
        green = 255 - im_pixel[x,y][1]
        blue = 255 - im_pixel[x,y][2]
        inverse_pixel[x,y] = (red, green, blue)
```

```python
inverse
```

```python
brightness = Image.new(mode="RGB", size=(lebar, tinggi))
brightness_pixel = brightness.load()
b = 100
for x in range(0, lebar):
    for y in range(0, tinggi):
        red = im_pixel[x,y][0] + b
        green = im_pixel[x,y][1] + b
        blue = im_pixel[x,y][2] + b
        brightness_pixel[x,y] = (red, green, blue)
```

```python
brightness
```

```python
contrast = Image.new(mode="RGB", size=(lebar, tinggi))
contrast_pixel = contrast.load()
g = 3
p = 105
for x in range(0, lebar):
    for y in range(0, tinggi):
        red = g * (im_pixel[x,y][0] - p) + p
        green = g * (im_pixel[x,y][1] - p) + p
        blue = g * (im_pixel[x,y][2] - p) + p
        contrast_pixel[x,y] = (red, green, blue)
```

```python
contrast
```

```python
threshold = Image.new(mode="RGB", size=(lebar, tinggi))
threshold_pixel = threshold.load()
for x in range(0, lebar):
    for y in range(9, tinggi):
        red = im_pixel[x,y][0]
        green = im_pixel[x,y][1]
        blue = im_pixel[x,y][2]
        gray = int((red+green+blue)/3)
        if gray >= 128 :
            gray = 255
        else : 
            gray = 0
        threshold_pixel[x,y] = (gray,gray,gray)
```
